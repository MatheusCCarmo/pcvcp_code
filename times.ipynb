```python
import os
import pandas as pd
import matplotlib.pyplot as plt


main_dir = "./output/symmetric"

instances = ["10.1", "10.2", "20.1", "20.2", "50.1", "50.2", "100.1", "100.2", "200.1", "200.2"]
algorithms = ["grasp", "grasp_vnd", "grasp_vns", "grasp_vns_vnd", "memetic_algorithm",
              "memetic_algorithm_vnd", "memetic_algorithm_vns", "memetic_algorithm_vns_vnd"]


average_durations = {algo: [] for algo in algorithms}

name_mapping = {
    "grasp": "GRASP",
    "grasp_vnd": "GRASP + VND",
    "grasp_vns": "GRASP + VNS",
    "grasp_vns_vnd": "GRASP + VNS + VND",
    "memetic_algorithm": "Algoritmo Memtico",
    "memetic_algorithm_vnd": "Algoritmo Memtico + VND",
    "memetic_algorithm_vns": "Algoritmo Memtico + VNS",
    "memetic_algorithm_vns_vnd": "Algoritmo Memtico + VNS + VND"
}

for instance in instances:
    for algorithm in algorithms:
        file_path = os.path.join("output", "symmetric", instance, f"{algorithm}.txt")
        
        if os.path.exists(file_path):
            df = pd.read_csv(file_path, delimiter=";", skiprows=0, header=0)
            avg_duration = df["duration(s)"].mean()
            average_durations[algorithm].append(avg_duration)
        else:
            print(f"Arquivo {file_path} no encontrado.")
            average_durations[algorithm].append(None)

plt.figure(figsize=(15, 10))

for algorithm, durations in average_durations.items():
    plt.plot(instances, durations, label=name_mapping[algorithm], marker='o')

plt.title("Comparao das mdias de tempo entre os algoritmos")
plt.ylabel("Mdia de Durao (s)")
plt.xlabel("Instncia")
plt.xticks(rotation=45)
plt.legend(title="Algoritmos", loc="upper left")
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.savefig("symmetric.png", format="png", dpi=300)
plt.show()
```
